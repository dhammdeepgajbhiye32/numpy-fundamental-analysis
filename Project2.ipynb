```python
# Here we will Generate temp data for 365 days and find the hottest and coldest day and 
# find the temperature deviation from average and identify outliers very High/Low temps.
```

```python
import numpy as np

# Step 1 : Generate random dataset of temperature for 365 days
temp_dataset = np.random.randint(5,49,size = 365)
print(f"Temperature Dataset for 365 days : \n {temp_dataset}")

# Step 2 : Find the hottest and coldest day
hottest_day = np.max(temp_dataset)
coldest_day = np.min(temp_dataset)
print(f"Hottest Day Temperature : {hottest_day}")
print(f"Coldest Day Temperature : {coldest_day}")

# Step 3 : Find the average temperature
avg_temp = np.mean(temp_dataset)
print(f"Average Temperature : {round(avg_temp, 2)}")

# Step 4 : Find the temperature deviation from average
temp_deviation = temp_dataset - avg_temp
print(f"Temperature Deviation from Average : \n {np.round(temp_deviation,2)}")

# Step 5 : Identify outliers (very High/Low temps)  

# For lowest outliers, we consider temperature is below 10 degrees
low_outliers = temp_dataset[temp_dataset < 10]
print(f"Low Temperature Outliers (Below 10 degrees) : \n {low_outliers}")
total_low_outliers = len(low_outliers)
print(f"Total Low Temperature Outliers : {total_low_outliers}") 

# For highest outliers, we consider temperature is above 40 degrees
high_outliers = temp_dataset[temp_dataset > 40]
print(f"High Temperature Outliers (Above 40 degrees) : \n {high_outliers}")
total_high_outliers = len(high_outliers)
print(f"Total High Temperature Outliers : {total_high_outliers}")

# End of the Project
```
